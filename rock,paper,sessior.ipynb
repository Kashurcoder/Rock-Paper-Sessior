```python
# game settings
intitaial = 0
rounds = int(input("enter the number of rounds:"))

points = 0
while intitaial<rounds:
  intitaial+=1
  import random
  list = ['rock','paper','sessior']
  user_input= input("enter your option:")

  computer_choice = random.choice(list)
  # win time
  if user_input == 'rock' and computer_choice== 'sessior':
    print("computer:",computer_choice)
    print("you won")
    points+=10
    print('your points',points)
  elif user_input == 'paper' and computer_choice== 'rock':
    print("computer:",computer_choice)
    print("you won")
    points+=10
    print('your points',points)

  elif user_input == 'sessior' and computer_choice== 'paper':
    print("computer:",computer_choice)
    print("you won")
    points+=10
    print('your points',points)

# lose time
  elif user_input == 'sessior' and computer_choice== 'rock':
    print("computer:",computer_choice)
    print("you lost")
    points-=10
    print('your points',points)
  elif user_input == 'rock' and computer_choice== 'paper':
    print("computer:",computer_choice)
    print("you lost")
    points-=10
    print('your points',points)

  elif user_input == 'paper' and computer_choice== 'sessior':
    print("computer:",computer_choice)
    print("you lost")
    points-=10
    print('your points',points)
  # both(exceptions)
  elif user_input == 'paper' and computer_choice== 'paper':
    print("computer:",computer_choice)
    print("tie")
    #  points-=10
    print('your points',points)

  elif user_input == 'sessior' and computer_choice== 'sessior':
    print("computer:",computer_choice)
    print("tie")
    #  points-=10
    print('your points',points)

  elif user_input == 'rock' and computer_choice== 'rock':
    print("computer:",computer_choice)
    print("tie")
    #  points-=10
    print('your points',points)








if points >0:
  print("you won with",points,"points")
else:
  print("you lost by",points,"points")



```
